Normal Distribution notation: X ~ N(mean, variance):Variable X follows a normal distribution with mean mu and variance sigma ^2

Appears in nature - size of fully grown male lion follows normal Distribution
Distinct Characteristics : 
1. bell shaped so majority of the data is centred around the mean so values further from the mean are less likely to occur
2. graph is symmetric with regards to the mean so values equally far away in opposing directions are equally likely

Expected value = mean
Variance is sigma ^2, usually given when distribution is defined - can be deduced by expected value
Var(X) = E(X^2) - E(X)^2
68% occur within 1 std of mean
95% occur within 2 std of mean
99.7% occur within 3 std of mean
=> outliers extremly rare in Normal Distribution

Standardising
Special Transformation where w emake expected value equal zero and the variance equal to 1: E(X) = 0 & Var(X)=1 
Distribution we get after standardising any normal distribution is called a Standard Normal Distribution
In addition to the 68/95/99.7 rule we get CDF -> Z score table

Want to move the graph left or right until the mean equals zero y=f(x-mu)
Want the standard deviation to be 1 y = f(x-mu/sigma)

Z ~ N(0,1)
Y ~ N(mu, sigma^2) => Transformation for standardising Z = (Y-mu) / sigma

In Standard Normal Distribution if y = mu + 2.3 sigma then Z = 2.3

Normal distribution requires a lot of data, if data is limited then outliers drastically affect 

Smaller samples - Students T distribution


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics

# Plot between -10 and 10 with .001 steps.
x_axis = np.arange(-20, 20, 0.01)

# Calculating mean and standard deviation
mean = statistics.mean(x_axis)
sd = statistics.stdev(x_axis)

plt.plot(x_axis, norm.pdf(x_axis, mean, sd))
plt.show()
```

```python
# importing numpy as np
import numpy as np

# importing pyplot as plt
import matplotlib.pyplot as plt

# position
pos = 0

# scale
scale = 10

# size
size = 10000


# random seed
np.random.seed(10)

# creating a normal distribution data
values = np.random.normal(pos, scale, size)

# plotting histograph
plt.hist(values, 100)

# plotting mean line
plt.axvline(values.mean(), color='k', linestyle='dashed', linewidth=2)

# showing the plot
plt.show()
```
